```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo
load_file = "dados_compras.json"
purchase_file = pd.read_json(load_file, orient = "records")
purchase_file.head(10)
```

## Informações Sobre os Consumidores

```python
# Implemente aqui sua solução
df_purchase = pd.DataFrame(purchase_file)
unique_logins = df_purchase['Login'].unique()

total_logins = 0

for i in unique_logins:
    total_logins += 1

print('Número total de consumidores: ',total_logins)
```

## Análise Geral de Compras

```python
# Implemente aqui sua solução
unique_itens = len(df_purchase['Item ID'].unique())
mean_price = round(df_purchase['Valor'].mean(),2)
total_item = df_purchase.shape[0]
total_value = df_purchase['Valor'].sum()

df_analise_compras = pd.DataFrame({"Número de itens exclusivos": unique_itens, "Preço média de compra": mean_price, "Número total de compras": total_item, "Faturamento":[total_value]})
df_analise_compras
```

## Análise Demográfica

```python
df_dados_clientes = df_purchase[['Login','Idade', 'Sexo']]
df_dados_clientes = df_dados_clientes.drop_duplicates()
```

```python
count = df_dados_clientes['Sexo'].value_counts()

```

```python
percent = round((count / total_logins) * 100,1)
percent
```

```python
df_analise_demografica = pd.DataFrame({"Sexo": count, "Porcentagem":percent})
df_analise_demografica
```

## Informações Demográficas Por Gênero

```python
# Implemente aqui sua solução
numero_compras = df_purchase.groupby(['Sexo']).count()['Valor']
```

```python
media_valor_compra=round(df_purchase.groupby(['Sexo']).mean()['Valor'],2)
```

```python
soma_compra = df_purchase.groupby(['Sexo']).sum()['Valor']
```

```python
df_analise_demografica_genero = pd.DataFrame({"Número de compras": numero_compras,"Valor médio de compras":media_valor_compra, "Valor total":soma_compra})
df_analise_demografica_genero
```

## Informações Demográficas Por Faixa Etária

```python
faixas = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999]
labels = ["Menos de 10", "10 a 14", "15 a 19", "20 a 24", "25 a 29", "30 a 34", "35 a 39", "Mais de 40"]

purchase_file["Range de Idades"] = pd.cut(purchase_file["Idade"], faixas, labels=labels)
```

```python
purchase_file.head()
```

```python
# Cálculos básicos
#contagem de pessoas em cada faixa etaria
age_demographics_count = purchase_file["Range de Idades"].value_counts() 
#aplicação da função média no valor da compra agrupando por faixa etaria
age_demographics_average_item_price = round(purchase_file.groupby(["Range de Idades"]).mean()["Valor"],2)
#aplicação da função soma no valor das compras de cada faixa etaria
age_demographics_total_item_price = purchase_file.groupby(["Range de Idades"]).sum()["Valor"]
#calculando a porcentagem de cada faixa em relação ao total de compradores
age_demographics_percent = round((age_demographics_count / total_logins) * 100,2)

# Dataframe para os resultados
age_demographics = pd.DataFrame({"Contagem": age_demographics_count, "%": age_demographics_percent, "Valor Unitario": age_demographics_average_item_price, "Valor Total de Compra": age_demographics_total_item_price})
age_demographics
```

## Consumidores Mais Populares (Top 5)

```python
# Implemente aqui sua solução
top5_valor_total = purchase_file.groupby(["Login"]).sum()["Valor"]
top5_numero_compras = purchase_file.groupby(["Login"]).count()["Valor"]
top5_media_preco = round(purchase_file.groupby(["Login"]).mean()["Valor"],2)

df_top5_login = pd.DataFrame({"Número de compras":top5_numero_compras,"Preço médio":top5_media_preco,"Valor total":top5_valor_total})
df_top5_login.sort_values("Valor total",ascending=False).head(5)

```

## Itens Mais Populares

```python
# Implemente aqui sua solução

item_compras = purchase_file.groupby(["Item ID"]).count()["Valor"]
item_valor = purchase_file.groupby(["Item ID"]).mean()["Valor"]
item_total = purchase_file.groupby(["Item ID"]).sum()["Valor"]

df_top_itens = pd.DataFrame({"Quantidade de compras":item_compras,"Valor do item":item_valor,"Total vendido":item_total})
df_top_itens.sort_values("Quantidade de compras", ascending=False)
```

## Itens Mais Lucrativos

```python
# Implemente aqui sua solução
df_top_itens.sort_values("Total vendido", ascending= False)
```
